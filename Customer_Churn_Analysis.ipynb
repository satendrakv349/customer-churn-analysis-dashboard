```python
#columns name in lower case and replace space with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
#create bins
tenure_labels = ['0-1 years ', '1-2 years', '2-4 years' , '4+ years']
df['tenure_group'] = pd.cut(df['tenure_months'], bins=4, labels=tenure_labels)
```

```python
#create monthly charges segment
monthly_charges_groups = [ 'Low', 'Medium', 'High']
df['monthly_charges_segment'] = pd.qcut(df['monthly_charges'], q=3, labels=monthly_charges_groups)
```
